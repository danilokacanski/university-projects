```python
import torch
from torch import nn

# Input: Temperature in 째C
X1 = torch.tensor([[10]], dtype=torch.float32) 
# Actual value: Temperature 째F
y1 = torch.tensor([[50]], dtype=torch.float32) 

# Input: Temperature in 째C
X2 = torch.tensor([[37.78]], dtype=torch.float32) 
# Actual value: Temperature 째F
y2 = torch.tensor([[100.0]], dtype=torch.float32) 

model = nn.Linear(1, 1)
loss_fn = torch.nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.0001)

# TODO: Replace X1, X2 with a single tensor X containing both rows
# TODO: Replace y1, y2 with a single tensor y containing both rows

for i in range(0, 100000):
    # Training pass (current: updates separately on X1 then X2)
    optimizer.zero_grad()
    outputs = model(X1)
    loss = loss_fn(outputs, y1)
    loss.backward()
    optimizer.step()

    optimizer.zero_grad()
    outputs = model(X2)
    loss = loss_fn(outputs, y2)
    loss.backward()
    optimizer.step()

    if i % 100 == 0: 
        print(model.bias)
        print(model.weight)

# TODO: Change training to use combined X and y for 150000 iterations
#       and keep printing model.bias and model.weight every 100 iterations

print("----")

# TODO: Put model into eval mode and, under no_grad, predict for measurements = [[37.5]] and print the prediction
```
