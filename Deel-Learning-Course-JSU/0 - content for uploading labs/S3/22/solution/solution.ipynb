```python
import torch
from torch import nn

# Input: Temperature in °C
X = torch.tensor([
    [10],
    [37.78]
], dtype=torch.float32)

# Actual value: Temperature °F
y = torch.tensor([
    [50],
    [100.0]
], dtype=torch.float32)

model = nn.Linear(1, 1)
loss_fn = torch.nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.0001)

# Change training to use combined X and y for 150000 iterations
# and keep printing model.bias and model.weight every 100 iterations
for i in range(150000):
    # Training pass
    optimizer.zero_grad()
    outputs = model(X)
    loss = loss_fn(outputs, y)
    loss.backward()
    optimizer.step()

    if i % 100 == 0: 
        print(model.bias)
        print(model.weight)

print("----")

# Put model into eval mode and, under no_grad, predict for measurements = [[37.5]] and print the prediction
measurements = torch.tensor([[37.5]], dtype=torch.float32)
model.eval()
with torch.no_grad():
    prediction = model(measurements)
    print(prediction)
```
