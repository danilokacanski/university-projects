```python
import sys
import torch
from torch import nn
import pandas as pd

df = pd.read_csv("student_exam_data.csv")

X = torch.tensor(
    df[["Study Hours", "Previous Exam Score"]].values,
    dtype=torch.float32
)
y = torch.tensor(df["Pass/Fail"], dtype=torch.float32)\
    .reshape((-1, 1))

# TODO: Replace the single Linear model with:
#       - one hidden Linear layer (2 → 10)
#       - one output Linear layer (10 → 1)
# TODO: Collect all parameters from both layers into a single list
# TODO: Update the training loop to:
#       - pass X through hidden layer
#       - apply sigmoid activation
#       - pass result through output layer
#       - compute loss and optimize

model = nn.Linear(2, 1)   # (to be replaced with hidden + output layers)
loss_fn = torch.nn.BCEWithLogitsLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.005)

for i in range(0, 500000):
    optimizer.zero_grad()
    outputs = model(X)  # (to be updated with hidden + sigmoid + output)
    loss = loss_fn(outputs, y)
    loss.backward()
    optimizer.step()

    if i % 10000 == 0:
        print(loss)

sys.exit()
model.eval()
with torch.no_grad():
    y_pred = nn.functional.sigmoid(model(X)) > 0.5
    y_pred_correct = y_pred.type(torch.float32) == y
    print(y_pred_correct.type(torch.float32).mean())
```
