```python
import sys
import torch
from torch import nn
import pandas as pd

df = pd.read_csv("student_exam_data.csv")

X = torch.tensor(
    df[["Study Hours", "Previous Exam Score"]].values,
    dtype=torch.float32
)
y = torch.tensor(df["Pass/Fail"], dtype=torch.float32)\
    .reshape((-1, 1))

model = nn.Sequential(
    nn.Linear(2, 10),
    nn.ReLU(),
    nn.Linear(10, 1)
)

loss_fn = torch.nn.BCEWithLogitsLoss()
# TODO: Use Adam optimizer with learning rate 0.005 (we’ll switch from the previous setup)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

# TODO: Train with mini-batches instead of full-batch:
#       - Define num_entries = X.size(0) and batch_size = 32
#       - Loop for a number of epochs (e.g., 1000)
#       - Inside each epoch, iterate start in range(0, num_entries, batch_size)
#         and slice X[start:end], y[start:end]
#       - Zero grad → forward on the slice → loss → backward → step
#       - Optionally print loss every few epochs

for i in range(0, 500000):
    optimizer.zero_grad()
    outputs = model(X)
    loss = loss_fn(outputs, y)
    loss.backward()
    optimizer.step()

    if i % 10000 == 0:
        print(loss)

model.eval()
with torch.no_grad():
    outputs = model(X)
    y_pred = nn.functional.sigmoid(outputs) > 0.5
    y_pred_correct = y_pred.type(torch.float32) == y
    print(y_pred_correct.type(torch.float32).mean())
```
