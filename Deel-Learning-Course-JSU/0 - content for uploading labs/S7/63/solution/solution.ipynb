```python
import torch
from torch import nn
import pandas as pd

df = pd.read_csv("../starter/student_exam_data.csv")

# Create feature tensor X using "Study Hours" and "Previous Exam Score"
X = torch.tensor(
    df[["Study Hours", "Previous Exam Score"]].values,
    dtype=torch.float32
)
# Create target tensor y using "Pass/Fail", reshaped to (-1, 1)
y = torch.tensor(df["Pass/Fail"], dtype=torch.float32)\
    .reshape((-1, 1))

# Define a Linear model with 2 inputs and 1 output
model = nn.Linear(2, 1)
# Define a BCEWithLogitsLoss loss function
loss_fn = torch.nn.BCEWithLogitsLoss()
# Define an SGD optimizer with learning rate 0.005
optimizer = torch.optim.SGD(model.parameters(), lr=0.005)

# Train the model for 500,000 iterations
for i in range(0, 500000):
    # zero gradients
    optimizer.zero_grad()
    # forward pass
    outputs = model(X)
    # compute loss
    loss = loss_fn(outputs, y)
    # backward pass
    loss.backward()
    # optimizer step
    optimizer.step()

    if i % 10000 == 0:
        print(loss)

# Put the model in eval mode and compute predictions
model.eval()
with torch.no_grad():
    # apply sigmoid and threshold at 0.5
    y_pred = nn.functional.sigmoid(model(X)) > 0.5
    # compare to true labels
    y_pred_correct = y_pred.type(torch.float32) == y
    # print mean accuracy
    print(y_pred_correct.type(torch.float32).mean())
```
