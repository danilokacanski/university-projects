```python
import pandas as pd

df = pd.DataFrame({
    "loan_intent":["PERSONAL", "EDUCATION"],
    "person_income": [30000, 50000]
})
# One-hot encode the column `loan_intent` using pd.get_dummies
# Convert all values to float32 with .astype("float32")
# Show the head of the dataframe
df = pd.get_dummies(df, columns=["loan_intent"]).astype("float32")
df.head()
```

```python
import torch

# Convert the dataframe values to a torch.tensor of dtype float32
# Print the tensor
X = torch.tensor(df.values, dtype=torch.float32)
print(X)
```
