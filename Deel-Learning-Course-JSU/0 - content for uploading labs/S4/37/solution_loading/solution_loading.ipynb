```python
import sys
import pandas as pd
import torch
from torch import nn

# Load X_mean from "X_mean.pt"
X_mean = torch.load("X_mean.pt")
# Load X_std from "X_std.pt"
X_std = torch.load("X_std.pt")
# Load y_mean from "y_mean.pt"
y_mean = torch.load("y_mean.pt")
# Load y_std from "y_std.pt"
y_std = torch.load("y_std.pt")

model = nn.Linear(2, 1)
# Load the model state_dict from "model.pt"
model.load_state_dict(torch.load("model.pt"))
# Set the model to evaluation mode
model.eval()

X_data = torch.tensor([
    [5, 10000],
    [2, 10000],
    [5, 20000]
], dtype=torch.float32)

with torch.no_grad():
    prediction = model((X_data - X_mean) / X_std)
    print(prediction * y_std + y_mean)
```
