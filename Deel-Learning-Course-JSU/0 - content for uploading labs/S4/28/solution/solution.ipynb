```python
import pandas as pd

df = pd.read_csv("../starter/used_cars.csv")
df.head()
```

```python
price = df["price"]

# Remove the "$" character from price (string replace)
price = price.str.replace("$", "", regex=False)

# Remove commas from price (string replace)
price = price.str.replace(",", "", regex=False)

# Convert price to integers (astype)
price = price.astype(int)

df["price"] = price
df.head()
```

```python
# Print the mean of the 'price' column
print(df["price"].mean())
```

```python
# Create a DataFrame df_hyundai with rows where brand == "Hyundai"
df_hyundai = df[df["brand"] == "Hyundai"]
# Print the mean price of df_hyundai
print(df_hyundai["price"].mean())

# Create a DataFrame df_ford with rows where brand == "Ford"
df_ford = df[df["brand"] == "Ford"]
# Print the mean price of df_ford
print(df_ford["price"].mean())
```

```python
# Create one-hot encodings for the 'brand' column using pd.get_dummies(df["brand"])
pd.get_dummies(df["brand"])
```

```python
import numpy as np

# Print the correlation coefficients between 'model_year' and 'price' using np.corrcoef
print(np.corrcoef(df["model_year"], df["price"]))
```

```python
import matplotlib.pyplot as plt

df_filtered = df[df["price"] < 300000]

# Create a scatter plot of model_year (x) vs price (y) using plt.scatter
plt.scatter(df_filtered["model_year"], df_filtered["price"])
plt.show()
```
