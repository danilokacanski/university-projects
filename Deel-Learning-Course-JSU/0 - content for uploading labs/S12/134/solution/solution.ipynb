```python
import gradio as gr
from PIL import Image

# Import PIL.Image so we can convert the uploaded NumPy array to a PIL image
# Change the second parameter to accept image pixels (a NumPy array) instead of text
# Print the Python type and shape of the uploaded image for inspection
# Convert the NumPy array to a PIL image with Image.fromarray(...)
# Keep the return as a simple greeting string for now
def greet(name, image_pixels):
    print(type(image_pixels))
    print(image_pixels.shape)

    img = Image.fromarray(image_pixels)
    # img.save("uploaded.jpg")
    print(img)

    return "Hello " + name + "!"

demo = gr.Interface(
    fn=greet,
    inputs=[
        gr.Text(label="Please enter your name:"),
        # Replace this TextArea with an Image uploader: gr.Image()
        gr.Image()
    ],
    # Keep a single text output for the greeting
    outputs=gr.Text(label="Output:")
)
demo.launch()
```
