```python
import gradio as gr

# Extend the greet function to also accept a "message" parameter
# and return both the greeting and the message.
def greet(name, message):
    return "Hello " + name + "! " + message

# Update the Interface so that:
#   1. It uses both inputs: a text box for the name, and a text area for the message.
#   2. It outputs the combined greeting text.
demo = gr.Interface(
    fn=greet,
    inputs=[
        gr.Text(label="Please enter your name:"),
        gr.TextArea(label="Please enter your message:")
    ],
    outputs=gr.Text(label="Output:")
)

demo.launch()
```
