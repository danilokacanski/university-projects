```python
import sys
import torch
from torch.utils.data import DataLoader
from torch import nn
import torch.nn.functional as F

from torchvision.transforms import ToTensor
from PIL import Image

model = nn.Sequential(
    nn.Linear(784, 50),
    nn.ReLU(),
    nn.Linear(50, 50),
    nn.ReLU(),
    nn.Linear(50, 10)
)
model.load_state_dict(torch.load('../starter/mnist-model.pth', weights_only=True))

# Open the image file "4.png" with PIL
img = Image.open("../starter/4.png")

# Resize the image to 28x28
img.thumbnail((28, 28))

# Convert the image to grayscale ('L')
img = img.convert('L')

# Convert the image to a tensor with ToTensor() and reshape to (-1, 784)
t = ToTensor()
X = t(img).reshape((-1, 784))

# Print the tensor shape
print(X.shape)

# Run the model on the tensor and print softmax probabilities
outputs = model(X)
print(nn.functional.softmax(outputs, dim=1))
```
