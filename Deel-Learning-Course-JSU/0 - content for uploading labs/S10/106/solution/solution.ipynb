```python
import torch
from torch import nn

# Select device (CPU, CUDA, or MPS depending on availability)
device = torch.device("cpu")
if torch.cuda.is_available():
    device = torch.device("cuda")
elif torch.mps.is_available():
    device = torch.device("mps")

# Print the device being used
print("Running on device:", device)

# Create tensor `a` directly on the chosen device
a = torch.tensor([5, 7], device=device)
# Create tensor `b` on CPU and then move it to the chosen device
b = torch.tensor([3, 4])
b = b.to(device)

# Print the devices of both tensors
print(a.device)
print(b.device)

# Print the result of adding them together
print(a + b)
```
