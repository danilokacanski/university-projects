```python
import torch
from torch import nn

device = torch.device("cpu")
if torch.cuda.is_available():
    device = torch.device("cuda")
elif torch.mps.is_available():
    device = torch.device("mps")

print("Running on device:", device)

a = torch.tensor([5, 7], device=device, dtype=torch.float32)
b = torch.tensor([3, 4], device=device)

print(a + b)

# Define a simple model with one Linear layer and move it to the chosen device
model = nn.Sequential(
    nn.Linear(2, 10)
).to(device)

# Run the model on `a.reshape(1, 2)`
outputs = model(a.reshape(1, 2))

# Apply softmax to the outputs
probabilities = nn.functional.softmax(outputs, dim=1)

# Print the probabilities
print(probabilities)
print(probabilities[0, :].tolist())
```
