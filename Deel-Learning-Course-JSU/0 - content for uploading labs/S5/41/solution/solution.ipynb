```python
import pandas as pd

df = pd.read_csv("../starter/SMSSpamCollection", sep="\t", names=["type", "message"])

# Create a boolean column 'spam' where value is True if type == "spam"
df["spam"] = df["type"] == "spam"

# Drop the original 'type' column
df.drop("type", axis=1, inplace=True)

# Show the first few rows
df.head()
```

```python
# Print the number of SPAM messages
print("SPAM messages:")
print(len(df[df["spam"] == True]))
```

```python
# Print the number of non-SPAM messages
print("Not SPAM messages:")
print(len(df[df["spam"] == False]))
```
